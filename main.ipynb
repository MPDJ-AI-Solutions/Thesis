### Loading and displaying images
To properly load images from disk and displaying them we can use this function that loads files from disk into array and then displaying them with matplotlib library

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


def load_image(image_path: str) -> np.ndarray:
    image_array = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    plt.imshow(image_array, cmap='gray')
    plt.show()
    plt.close()
    return image_array
```

### Example code
This example show all images of area in different wavelength and those processed with mag1c algorithm. (as photos and array) 

```python
import glob

photos_paths_list = [file for file in glob.glob(TEST_CATALOG + test_set_description['id'][256] + "/*.tif")]

for image_path in photos_paths_list:
    try:
        image_array = load_image(image_path)
        print(image_array)
    except Exception as exception:
        print('Cannot read image: ', image_path)
```
